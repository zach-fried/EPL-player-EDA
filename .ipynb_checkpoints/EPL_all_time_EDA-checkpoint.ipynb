# Analyzing Jersey Numbers

```python
data['Jersey Number'].unique()
```

```python
data[data['Jersey Number'].isnull()]
```

```python
data = data.dropna(subset=['Jersey Number'])
```

```python
data['Jersey Number'].isnull().sum()
```

```python
data['Jersey Number'] = data['Jersey Number'].astype(int)
```

```python
data['Jersey Number'].unique()
```

### Creating a new feature in the original dataset called "Total Chances" that sums a players goals and assists

```python
data['Total chances'] = data['Goals'] + data['Assists']
```

```python
data['Total chances'].isnull().sum()
```

```python
jersey_num_means = data.groupby('Jersey Number').agg(np.mean)
```

```python
jersey_num_means['Goals per match'] = jersey_num_means['Goals per match'].fillna(0)
```

```python
jersey_num_means['Total chances'][jersey_num_means['Total chances'] > 0].plot(kind='barh', figsize=(20,30))
plt.text(x=18, y=44, s='Total Chances by Jersey Number', fontsize=22)
```

Unsurprisingly, numbers 9 & 10 produced the greatest number of chances (goals + assists). Oddly, number 66 produced a significant number of chances on average. I wonder if there's a way to dig into the players that wore number 66.

```python
data[data['Jersey Number'] == 66][['Name','Goals','Assists']]
```

And there's our answer - there is only a single player in the dataset that wore number 66 and it is none other than Alexander-Arnold of Liverpool. He is a great, young talent and does create a significant number of chances.
